The classification scheme is based on the event types defined by Dal Maso et al. (2005):
- Class I: Days when the growth and formation rate could be determined with a good confidence level
  - Class Ia: Very clear and strong particle formation events, with little or no pre-existing particles obscuring the newly formed mode
  - Class Ib: The rest of Class I events
- Class II: Days where the derivation of these parameters was not possible or the accuracy of the results was questionable
- Nonevent
